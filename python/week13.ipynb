```python
#내 생일 서울의 최고, 최저 기온 시각화
import csv
import matplotlib.pyplot as plt
f = open('temp.csv')
data = csv.reader(f)
next(data)
max_temp = []
min_temp = []
for row in data:
    if row[3]!='':
        if row[0].split('.')[1]=='1' and row[0].split('.')[2]=='25':
            min_temp.append(float(row[3]))
    if row[4]!='':
        if row[0].split('.')[1]=='1' and row[0].split('.')[2]=='25':
            max_temp.append(float(row[4]))

plt.plot(min_temp, color = 'skyblue', label='low')
plt.plot(max_temp, color = 'pink', label='high')
plt.legend()
plt.show()
```

```python
#최고 기온 시각화
import csv
import matplotlib.pyplot as plt
f = open('temp.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data:
    if row[4]!='':
        if row[0].split('.')[1]=='8':
            aug.append(float(row[4]))
        elif row[0].split('.')[1]=='1':
            jan.append(float(row[4]))
            
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```
